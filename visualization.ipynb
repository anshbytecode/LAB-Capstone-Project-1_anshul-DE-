```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
quantity_sold = [120, 90, 150, 60, 30]
revenue = [1200, 900, 1500, 600, 300]

days = pd.date_range(start='2025-09-01', periods=7)
daily_revenue = [1000, 1200, 900, 1500, 1700, 1600, 1800]
daily_temp = [28, 30, 27, 32, 31, 29, 33]
```

```python

plt.figure(figsize=(8,5))
plt.bar(products, quantity_sold, color='orange')
plt.title("Quantity Sold per Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python

plt.figure(figsize=(6,6))
plt.pie(revenue, labels=products, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title("Revenue Share by Product")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.barh(products, revenue, color='green')
plt.title("Revenue per Product")
plt.xlabel("Revenue")
plt.ylabel("Product")
plt.show()
```

```python

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.set_xlabel('Day')
ax1.set_ylabel('Revenue', color='tab:blue')
ax1.plot(days, daily_revenue, color='tab:blue', marker='o', label='Revenue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Avg Temp (°C)', color='tab:red')
ax2.plot(days, daily_temp, color='tab:red', marker='x', linestyle='--', label='Avg Temp')
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.tight_layout()
plt.title("Daily Revenue vs Average Temperature")
fig.legend(loc="upper left", bbox_to_anchor=(0.1,0.9))
plt.show()
```
